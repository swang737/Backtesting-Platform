```python
# parent strategy class
# be careful, overfitting since testing on same data --> just useful
class Strategy:
    def __init__(self, DataLoader, fee):
        self.loader = DataLoader
        self.fee = 0.001

    def entryIndicator(self):
        pass

    def exitIndicator(self):
        pass

    def anchorWindow(self):
        pass

    def floatWindow(self):
        pass
```

```python
class ARstrategy(Strategy):
    def __init__(self, DataLoader, fee):
        super().__init__(DataLoader, fee)
    
    def entryIndicator(self, prices):
        # checking to see if we make more than the 0.001 commission rate
        # pasting from fitted ARIMA model
        if (prices[-1] / prices[-2]) ** 0.3195  > 1.001:
            return 1
        elif (prices[-1] / prices[-2]) ** 0.3195 < 0.999:
            return 0
        else:
            return None

    def getPosition(self, currPos = []):
        # resetting current pos
        currPos = [0] * self.loader.nins

        for i in range(self.loader.nins):
            prices = self.loader.data.T[i]
            if self.entryIndicator(prices) == 1:
                currPos[i] = 99999999999999
            elif self.entryIndicator(prices) == 0:
                currPos[i] = -99999999999999
            else:
                currPos[i] == 0

        return currPos

    def anchorWindow(self):
        currPos = self.getPosition()

        while self.loader.t < self.loader.nt:
            currPos = self.getPosition(currPos)
            self.loader.stepTime()
            print(currPos)   
```

```python
StockData = np.loadtxt('../Data/StockData.txt')
```

```python
commission = 0.001
strat = ARstrategy(DataLoader(StockData, [38]), commission)
strat.loader.t = 1000
print(strat.loader.nt)
strat.anchorWindow()

# seems to be no entries or exits in the dataset (rip) likely stock 37 ar strat untradable
```
